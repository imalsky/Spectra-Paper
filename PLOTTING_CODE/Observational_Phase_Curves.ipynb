```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


plt.style.use('science')
```

```python

df = pd.read_csv('../DATA/HD189bo11.csv')

y_data = (df.bestfit_norm[::100] - 1) * 1e6

plt.scatter(df.phase[::100], y_data, s=1)
plt.ylim(0, max(y_data) * 1.2)
plt.xlabel('Orbital Phase')
plt.ylabel(r'F$_p$/F$_s$ (ppm)')
plt.title('HD 189733 b Best Fit Phase Curve')
```
